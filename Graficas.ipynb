```python
#Carguemos los datos
import pandas as pd
ShanghaiDat=pd.read_csv('Shanghai-2005-2025.csv')
ShanghaiDat=ShanghaiDat.drop(ShanghaiDat.columns[0], axis=1) #Eliminar la primera columna que tiene la etiqueta

print('Dimension del dataframe:',ShanghaiDat.shape)
print('Primer elemento del dataframe:',ShanghaiDat.iloc[0,0])
```

```python
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

lista_fechas = pd.to_datetime(ShanghaiDat.columns)
val = ShanghaiDat.iloc[0, :].to_numpy()

sns.set_theme(style="whitegrid")  # opciones: whitegrid, darkgrid, ticks

plt.figure(figsize=(12, 5))
sns.lineplot(x=lista_fechas, y=val, label='Datos Shangai')
plt.title('Serie temporal del indice financiero')
plt.xlabel('Fecha')
plt.ylabel('Precio')

plt.gca().xaxis.set_major_locator(mdates.YearLocator(2))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```
